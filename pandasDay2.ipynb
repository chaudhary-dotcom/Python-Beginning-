```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\user\OneDrive\Documents\DataScience\Python\supermarket_sales - Sheet1.csv')
df.head()
```

```python
df[['Branch', 'City', 'Total']]
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head(3)
```

```python
df['Rank'] = df['Rating'].rank(method='min', ascending=False).astype(float)
df.head(5)
```

```python
total_revenue_per_city = df.groupby('City')['Total'].sum().sort_values(ascending=False)
print('\n Total revenue per City:\n', total_revenue_per_city)
```

```python
highest_revenue_per_city = df.groupby('City')['Total'].sum().idxmax()
print('\n Highest Revenue City\n', highest_revenue_per_city)
```

```python
total_revenue_per_product = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
print('\n Total Revenue per product\n', total_revenue_per_product)
```

```python
highest_revenue_product = df.groupby('Product line')['Total'].sum().idxmax()
print('\n Highest Revenue Product\n', highest_revenue_product)
```

```python
df[['Branch', 'City', 'Product line', 'Total']]
```

```python
highest_quantity_sold = df.groupby('Product line')['Quantity'].sum().sort_values(ascending=False)
print('\n Highest Quantity Sold\n', highest_quantity_sold)
```

```python
highest_sold_quantity = df.groupby('Product line')['Quantity'].sum().idxmax()
print('\nHighest Sold Quantity\n', highest_sold_quantity)
```

```python
average_revenue_per_product = df.groupby('Product line')['Total'].mean()
print('\nAverage Revenue Per City\n', average_revenue_per_product)
```

```python
#concatenate two data frames using the following conditions
#create two data frames using following dictionaries
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMW', 'Audi'], 'Price': [23845, 171995, 135925, 71400]}
carDf1 = pd.DataFrame.from_dict(GermanCars)

japneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi'], 'Price': [29995, 23600, 61500, 58900]}
carDf2 = pd.DataFrame.from_dict(japneseCars)

carsDf = pd.concat([carDf1, carDf2], keys = ['Germany', 'Japnese'])
carsDf
```

```python
#Merge two data frames and append new data frame as a new column
car_price = {'Company': ['Toyota', 'Honda', 'BMW', 'Audi'], 'Price': [23845, 17995, 135925, 71400]}
carPriceDf = pd.DataFrame.from_dict(car_price)

car_horsepower = {'Company': ['Toyota', 'Honda', 'BMW', 'Audi'], 'horsepower': [141, 80, 182, 160]}
carHorsepowerDf = pd.DataFrame.from_dict(car_horsepower)

carsDf = pd.merge(carPriceDf, carHorsepowerDf, on = 'Company')
carsDf
```

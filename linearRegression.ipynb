```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
```

```python
df = fetch_california_housing()
```

```python
df
```

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns = df.feature_names
dataset
```

```python
# Independent features and dependent fetures
x = dataset
y = df.target
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_trian, y_test = train_test_split(
    x, y, test_size=0.30, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
# cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train, y_trian)
```

```python
mse = cross_val_score(regression, X_train, y_trian, scoring='neg_mean_squared_error', cv=5)
```

```python
np.mean(mse)
```

```python
# predication
reg_pred = regression.predict(X_test)
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred - y_test, kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(reg_pred, y_test)
score
```

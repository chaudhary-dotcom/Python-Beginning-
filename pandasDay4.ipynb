```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Costumer_id': [101, 102, 103, 104, 105, 106],
    'Product': ['A', 'B', 'C', 'A', np.nan, 'C'],
    'Amount': [250, np.nan, 150, 400, 200, np.nan],
    'Discount': [10,15,np.nan, 5, 20,10]
}

df = pd.DataFrame(data)
df
```

```python
#Fill missing values in amount
df['Amount'].fillna(df['Amount'].mean(), inplace = True)
df
```

```python
#fill missing values in discount with 0
df['Discount'].fillna(0, inplace=True)
df
```

```python
df.dropna(subset=['Product'], inplace=True)
df
```

```python
df['Sales'] = [1000, 500, 1200, 700, 300]
df['Profit'] = [100, 50, 120, 70, 30]

df['Revenue'] = df['Sales'] * df['Profit']
df
```

```python
Average_profit_per_product = df.groupby('Product')['Profit'].mean()
Average_profit_per_product
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# Define matrix
A = np.array([[4, -2], 
              [1,  1]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Plot original eigenvectors
origin = np.array([0, 0])  # Origin for quiver plot

plt.quiver(*origin, eigenvectors[0, 0], eigenvectors[1, 0], color='r', angles='xy', scale_units='xy', scale=1, label='Eigenvector 1')
plt.quiver(*origin, eigenvectors[0, 1], eigenvectors[1, 1], color='b', angles='xy', scale_units='xy', scale=1, label='Eigenvector 2')

# Set plot limits
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.legend()
plt.title("Eigenvectors of Matrix A")
plt.show()
```

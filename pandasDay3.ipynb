```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\user\OneDrive\Documents\DataScience\Python\supermarket_sales - Sheet1.csv')
df.head(3)
```

```python
df['Revenue']  = df['Quantity'] * df['Unit price']
result = df.groupby(['City', 'Product line'])['Revenue'].sum().sort_values(ascending=False)
result
```

```python
#unique city
num_unique = df['City'].nunique()
unique = df['City'].unique()
print('Total city:',num_unique)
print('Name of city:',unique)

#uique Product
num_product = df['Product line'].nunique()
print('Number of products: ', num_product)

product_name = df['Product line'].unique()
print('Product name: ', product_name)

```

```python
highest_rating_product = df.groupby('Product line')['Rating'].max().sort_values(ascending=False)
highest_rating_product
```

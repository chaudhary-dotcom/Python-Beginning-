Mean(): fill null values in column using mean formula

```python
df=pd.read_csv("/data.csv")
a=df["Calories"].mean()
df.fillna({"Calories":a},inplace=True)
df.head(30)
```

```python
b=df["Calories"].median()
df.fillna({"Calories":b},inplace=True)
df.head(30)
```

Cleaning Data of wrong format

```python
data={
    "Date":{
        "0":"2022/2/3",
        "1":"2022/3/3",
        "2":"5/2/2022",
        "3":"6/3/2022",
        "4":123 ,
        "5":"2022/4/4"

    },
     "Duration": {
        "0":60,
        "1":60,
        "2":60,
        "3":60,
        "4":45,
        "5":45

    },
    "Pulse":{
        "0":110,
        "1":117,
        "2":103,
        "3":109,
        "4":117,
        "5":102
    },
    "Maxpulse": {
        "0":130,
        "1":145,
        "2":135,
        "3":282,
        "4":406,
        "5":300
    },
}
df=pd.DataFrame(data)
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.to_string()
```

```python
df.dropna(subset=['Date'], inplace=True)
df
```

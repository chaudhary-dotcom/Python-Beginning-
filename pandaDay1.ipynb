```python
#concatenate two data frames using the following conditions
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMW', 'Audi'], 'Price': [23845, 171995, 135925, 71400]}
carsDf1 = pd.DataFrame.from_dict(GermanCars)

japneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi'], 'Price': [29995, 23600, 61500, 58900]}
carsDf2 = pd.DataFrame.from_dict(japneseCars)

carsDf = pd.concat([carsDf1, carsDf2], keys=['Germany', 'Japan'])
carsDf
```

```python
#Merge two dataframes 
car_price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925, 71400]}
car_pricedf = pd.DataFrame.from_dict(car_price)

car_horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141,80,182,160]}
car_horsepowerdf = pd.DataFrame.from_dict(car_horsepower)
carsDf = pd.merge(car_pricedf, car_horsepowerdf, on = 'Company')
carsDf
```

```python
#Applying function
def categorize_salary(salary):
    if salary < 60000:
        return 'Low'
    elif 60000 <= salary <= 75000:
        return 'Medium'
    else:
        return 'High'
    
df['Salary_level'] = df['Salary'].apply(categorize_salary)

df
```

```python
df['Joining_date'] = ['2025-5-30', '2024-8-14', '2023-6-12', '2022-1-11', '2019-04-21']
df['Joining_date'] = pd.to_datetime(df['Joining_date'])
df['Joining_year'] = df['Joining_date'].dt.year
df['Joining_month'] = df['Joining_date'].dt.month_name()
df['Joining_day'] = df['Joining_date'].dt.day_name()
df
```

```python
data1 = {
    'Date': ['2025-01-01', '2025-01-02', '2025-01-01', '2025-01-02', '2025-03-03'],
    'Store': ['A', 'A', 'B', 'A', 'B'],
    'Sales': [100,200,300, 400, 500],
    'Profit': [50,60,70, 80, 90]
}
#datetime
df1 = pd.DataFrame(data1)
df1['Date'] = pd.to_datetime(df1['Date'])
df1

#groupby
grouped = df1.groupby(['Store', 'Date']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
grouped

```

```python
pivot_table = df1.pivot_table(values=['Sales', 'Profit'], index='Date', columns='Store', aggfunc='sum')
pivot_table
```

```python
data = {
    'Costumer': ['Alice', 'Bob', 'Smith', 'David', 'Frank', 'Eve'],
    'Age': [25, np.nan, 35, 40, np.nan, 28],
    'Purchase_Amount': [100, 200, np.nan, 500, np.nan, 800]
}

df_1 = pd.DataFrame(data)
df_1
```

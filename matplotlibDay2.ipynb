```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

#plot 1:
plt.subplot(1,2,1)
plt.plot(x, y)

#plot:2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
x = np.array([0,1,2,3,4])
y = np.array([2,4,6,8,10])

# plot 1:
plt.subplot(2,1,1)
plt.plot(x, y, color='r', label='Plot 1', marker ='o', linestyle='--', markersize=5)
plt.legend()

# plot-2
x = np.array([1,2,3,4,5])
y = np.array([1,5,2,5,8])
plt.subplot(2,1,2)
plt.plot(x, y)

plt.show()
```

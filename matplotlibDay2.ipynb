```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)
scatter_x = np.random.rand(20) * 10
scatter_y = np.random.rand(20) * 10
bar_x = ['A', 'B', 'C', 'D', 'E']
bar_y = [5, 7, 3, 8, 6]
z = np.linspace(0, 10, 100)
x_3d = np.sin(z)
y_3d = np.cos(z)

# Create figure and 22 grid
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot (2D)
axs[0, 0].plot(x, y, color='blue', label='Sine Wave')
axs[0, 0].set_title("Line Plot")
axs[0, 0].legend()

# Scatter plot (2D)
axs[0, 1].scatter(scatter_x, scatter_y, color='red', marker='o')
axs[0, 1].set_title("Scatter Plot")

# Bar plot (2D)
axs[1, 0].bar(bar_x, bar_y, color='green')
axs[1, 0].set_title("Bar Plot")

# 3D Plot
ax_3d = fig.add_subplot(2, 2, 4, projection='3d')
ax_3d.plot3D(x_3d, y_3d, z, color='purple')
ax_3d.set_title("3D Line Plot")

# Adjust layout for proper spacing
plt.tight_layout()
plt.show()
```

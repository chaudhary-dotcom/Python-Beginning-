**Marker**

```python
#marker
import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='_')
plt.show()
```

```python
plt.plot(ypoints,'o:r')
plt.show()
```

```python
#marker size: ms

plt.plot(ypoints, marker='o',ms=20)
plt.show()
```

```python
#marker, size and markeredgecolor(mec)

plt.plot(ypoints, marker ='*',ms=20,mec='r')
plt.show()
```

```python
#markerfacecolor: mfc

plt.plot(ypoints,marker='*',ms=20,mfc='r')
plt.show()
```

```python
#Using both mec: marker edge color and mfc: marker face color
plt.plot(ypoints, marker='s',ms=20, mec='r',mfc='y')
plt.show()
```

**Labels**

```python
#labels
import matplotlib.pyplot as plt
import numpy as np

x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x, y)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
#Title
plt.plot(x,y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

**Font Properties for Title and Labels**

```python
font1 = {'family':'serif', 'color':'blue', 'size':20}
font2 = {'family':'serif', 'color':'darkred', 'size':15}

plt.title("Sports Watch Data", fontdict=font1)
plt.xlabel("Average Pulse", fontdict =font2)
plt.ylabel("Calorie Burnage", fontdict =font2)

plt.plot(x, y)
plt.show()
```

```python
#loc: locate or align the title left, right and center
plt.title("Sports Watch Data", loc='left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

**Grid Lines**

```python
#Adding Gridlines using grid() function
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
#axis of grid line x-axis
plt.title("Sports Watc Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.grid(axis='x')
plt.show()
```

```python
#axis of grid line Y-axis
plt.title("Sports Watc Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.grid(axis='y')
plt.show()
```

```python
#Line properties: color, linestyle, linewidth

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.grid(color='green', linestyle='--',linewidth=0.5)
plt.show()
```

```python
#Scatter

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.title("Car Passing", fontsize=20)
plt.xlabel("Age Of the Car")
plt.ylabel("Speed of the car")
plt.show()
```

```python
#day one, the age and the speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y, color='hotpink')


#day two, the age and speed of 15 cars:
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color='#88c999')

plt.show()
```

```python
#color each dot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x, y, c=colors)
plt.show()
```

```python
#color-map: cmap
#viridis: it is built-in colormaps available in matplotlib
#Colorbar: colorbar()

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors=np.array([0,10,20,30,40,45,50,55,60,70,80,90,100])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y, c=colors, cmap='viridis', s=sizes)
plt.colorbar()
plt.show()
```

```python
#alpha: adjust the transperrancy of the dots

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y, s=sizes, alpha=0.5)

plt.show()
```

```python
#Create random arrays with 100 values for x-points, y-points, colors and sizes:

x=np.random.randint(100, size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))

plt.scatter(x,y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()

plt.show()
```

##Horizontal Bars

```python
#barh(): if you want the bars to be displayed horizontally instead of vertically

x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.barh(x,y, color='r')
plt.show()
```

```python
#width of bar
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x, y, width=0.1)
plt.show()
```

```python
#For horizontal bars use height instead of width
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.barh(x, y, height=0.1)
plt.show()
```

```python
#Pie charts: pie()
y=np.array([35,25,25,15])

plt.pie(y)
plt.show()
```

```python
#Adding label
y=np.array([35,25,25,15])
mylabels = ['APPLE','BANANA','CHERRIES','DATES']

plt.pie(y, labels=mylabels)
plt.show()
```

```python
#Startangle: The default startangle is x-axis, but you can change by specifying a startangle parameter
y=np.array([35,25,25,15])
mylabels=['Apples','Bananas','Cherries', 'Dates']

plt.pie(y, labels=mylabels, startangle=90)
plt.show()
```

```python
#Explode
#Shadow
y=np.array([35,25,25,15])
mylabels=['Apple','Banana','Cherries','Dates']
myexplode=[0.2,0,0,0]

plt.pie(y,labels=mylabels, explode=myexplode, shadow=True)
plt.show()
```

```python
#Color
#legend
y=np.array([35,25,25,15])
mylabels=['Apple','Banana','Cherries','Dates']
mycolor=['black','hotpink','#1f2f5c', 'gray']

plt.pie(y, labels=mylabels, colors=mycolor)
plt.legend()
plt.show()
```

```python
#Title parameter to legend function

y=np.array([35,25,25,15])
mylabels=['Apple','Banana','Cherries','Dates']
mycolor=['black','hotpink','#1f2f5c', 'gray']

plt.pie(y, labels=mylabels, colors=mycolor)
plt.legend(title ='Four Fruits')
plt.show()
```

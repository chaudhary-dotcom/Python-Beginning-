```python
#Student Grade Syetem
class Student:
    def __init__(self, name, student_id):
        self.name = name 
        self.student_id = student_id
        self.grades = []

    def add_grade(self, score):
        if 0 <= score <= 100:
            self.grades.append(score)
            print(f'{score} added for {self.name}')
        else:
            print('Invalid grade! please enter a score between 0 and 100.')
        
    def get_average(self):
        if not self.grades:
            print(f"No grades available for {self.name}")
            return None
        return sum(self.grades) / len(self.grades)
    
    def print_transcipt(self):
        print(f"Transcipt for {self.name} (ID: {self.student_id}):")
        if not self.grades:
            print('No grades available')
        
        else:
            print('Grades:', ','.join(map(str, self.grades)))
            print(f"Average Grades: {self.get_average():.2f}")
        
s1 = Student('Chilwal', '56jh')
s1.add_grade(76)
s1.add_grade(87)
print(s1.get_average())
s1.print_transcipt()
```

```python
class Room:
    def __init__(self, room_number, room_type, price_per_night):
        """Initialize a room with details like number, type, price, and availability."""
        self.room_number = room_number
        self.room_type = room_type  # e.g., "Single", "Double", "Luxury"
        self.price_per_night = price_per_night
        self.is_booked = False  # Default availability is False

    def book(self):
        """Mark the room as booked."""
        if not self.is_booked:
            self.is_booked = True
            print(f"Room {self.room_number} has been successfully booked.")
        else:
            print(f"Room {self.room_number} is already booked!")

    def checkout(self):
        """Mark the room as available."""
        if self.is_booked:
            self.is_booked = False
            print(f"Room {self.room_number} is now available.")
        else:
            print(f"Room {self.room_number} was not booked.")

    def __str__(self):
        """Return a string representation of the room details."""
        status = "Available" if not self.is_booked else "Booked"
        return f"Room {self.room_number} | Type: {self.room_type} | Price: ${self.price_per_night}/night | Status: {status}"


class Hotel:
    def __init__(self, name):
        """Initialize the hotel with a name and an empty list of rooms."""
        self.name = name
        self.rooms = []  # List to store Room objects
        self.revenue = 0  # Track revenue generated by bookings

    def add_room(self, room):
        """Add a room to the hotel."""
        self.rooms.append(room)

    def display_available_rooms(self):
        """Display all available rooms."""
        print(f"\nAvailable Rooms in {self.name}:")
        available_rooms = [room for room in self.rooms if not room.is_booked]
        if not available_rooms:
            print("No rooms are currently available.")
        else:
            for room in available_rooms:
                print(room)

    def book_room(self, room_type):
        """Book the first available room of a specified type."""
        for room in self.rooms:
            if room.room_type == room_type and not room.is_booked:
                room.book()
                self.revenue += room.price_per_night  # Add to revenue
                return
        print(f"No {room_type} rooms are currently available for booking.")

    def checkout_room(self, room_number):
        """Mark a room as available based on its room number."""
        for room in self.rooms:
            if room.room_number == room_number:
                room.checkout()
                return
        print(f"Room {room_number} does not exist in {self.name}.")

    def display_revenue(self):
        """Display total revenue generated by the hotel."""
        print(f"\nTotal Revenue for {self.name}: ${self.revenue}")


# Example Usage
hotel = Hotel("Dream Stay")

# Adding rooms
hotel.add_room(Room(101, "Single", 100))
hotel.add_room(Room(102, "Double", 150))
hotel.add_room(Room(103, "Luxury", 300))
hotel.add_room(Room(104, "Single", 100))

# Display available rooms
hotel.display_available_rooms()

# Book rooms
hotel.book_room("Single")
hotel.book_room("Luxury")
hotel.book_room("Single")

# Try to book a room that is already booked
hotel.book_room("Single")

# Checkout a room
hotel.checkout_room(101)

# Try to checkout a room that isn't booked
hotel.checkout_room(104)

# Display available rooms again
hotel.display_available_rooms()

# Display total revenue
hotel.display_revenue()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
#generate data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# create 3d plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection ='3d')
ax.plot_surface(X, Y, Z, cmap ='viridis')

#lables
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z AXis')
ax.set_title('3D surface plot')
# plt.bar(x, y)

# plt.plot(x, y)
plt.show()
```

```python
#bar plot
a = ['A', 'B', 'C']
b = [10,20,15]
plt.subplot(3, 2,1)
plt.bar(a, b, color='blue')
plt.xlabel('A')
plt.ylabel('B')
plt.title('Bar plot')
plt.tight_layout()

# plot-2:
plt.subplot(3,2,2)
data = [1,2,2,3,3,3,4,4,5]
plt.hist(data, bins=5,color='green')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.tight_layout()

# plot:3
plt.subplot(3,2,3)
labels =['apple', 'banana', 'orange']
sizes = [40,35,25]
plt.pie(sizes, labels=labels, autopct = '%1.1f%%')
plt.title('Pie chart')
plt.tight_layout()

# boxplot
plt.subplot(3,2,4)
data = [7,8,6,9,10,15,14,12]
plt.boxplot(data)
plt.title('Box plot')
plt.tight_layout()

# area plot
plt.subplot(3,2,5)
x = [1,2,3,4,5]
y= [1,4,9,16,25]
plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.plot(x, y, color='blue')
plt.title('Area plot')
plt.show()
```

```python
# contour plot
x = np.linspace(-5,5,100)
y = np.linspace(-5,5, 100)

X, Y =np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)
plt.contour(X, Y, Z)
plt.title('Contour plot')
# plt.scatter(x, y)
plt.show()
```

```python
# data
t = np.linspace(0,20,100)
x = np.sin(t)
y = np.cos(t)
z = t

# plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='3D helix')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
ax.set_title('3d line plot')
ax.legend()
# plt.tight_layout()
plt.show()
```

```python
t = np.linspace(0, 20,100)
x = np.sin(t)
y = np.cos(t)
z = t**2

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot(x, y, z, color='r', label='3d helix')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.title('3d ploting')
# plt.tight_layout()
plt.legend()
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
tips.info()
# sns.set_style('whitegrid')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Seaborn scatter plot')
plt.show()
# print(tips)
```

```python
flights = sns.load_dataset('flights')
sns.lineplot(x='year', y='passengers', data=flights)
plt.show()
```

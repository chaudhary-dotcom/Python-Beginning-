```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = [1,2,3,4,5]
b= [10,20,30,40,50]
plt.plot(a, b)
plt.show()
```

```python
plt.plot(a, b, color = 'red', linestyle='--', marker='o', markersize=8, label='Line 1')
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.title('Customized plot')
plt.legend()
plt.show()
```

```python
plt.bar(a, b, color='blue')
plt.xlabel('category')
plt.ylabel('Values')
plt.title('Bar chart')
```

```python
plt.scatter(a, b, color='red', marker='o')
plt.xlabel('a label')
plt.ylabel('b label')
plt.title('Scatter plot')


plt.show()
```

```python
#subplots
fig, axes = plt.subplots(1,2, figsize=(10,4))
axes[0].plot(a,b, color='green')
axes[0].set_title('line plot')

categories = ['A', 'B', 'C', 'D']
values = [5,7,3,8]
axes[1].bar(categories, values, color='orange')
axes[1].set_title('bar chart')
plt.tight_layout()
plt.show()
```
